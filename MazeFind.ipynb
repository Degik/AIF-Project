```python
import gym
import minihack
import numpy as np
import matplotlib.pyplot as plt
import IPython.display as display
import utils
import algorithms as algo
```

```python
env = gym.make("MiniHack-MazeWalk-Mapped-45x19-v0", observation_keys=("chars", "pixel"))
state = env.reset()
env.render()
plt.imshow(state["pixel"][25:325, 200:1025])
game_map = state["chars"]
game = state["pixel"]
```

```python
start = utils.get_player_location(game_map)
target = utils.get_target_location(game_map)
print(start)
print(target)
```

```python
path = algo.a_star(game_map, start, target, utils.manhattan_distance)
print(path)
```

```python
actions = utils.actions_from_path(start, path[1:])
print(actions)
```

```python
import time
image = plt.imshow(game[23:325, 250:1025])
for action in actions:
    s, _, _, _ = env.step(action)
    display.display(plt.gcf())
    display.clear_output(wait=True)
    image.set_data(s["pixel"][25:325, 250:1025])
```
